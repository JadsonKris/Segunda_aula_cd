```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/houses_to_rent_v3.txt')
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribui巽達o de 叩reas dos im坦veis')
plt.hist(df['area'])
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta e valor da gorjeta')
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta e valor da gorjeta')
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta e valor da gorjeta pelo g棚nero')
sns.scatterplot(x="total_bill", y="tip", hue = 'sex', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta e valor da gorjeta pelo dia da semana')
sns.scatterplot(x="total_bill", y="tip", hue = 'day', data=tips)
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", col = 'day', data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue = 'sex', data=tips)
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", hue = 'size', data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Rela巽達o entre valor da conta, valor da gorjeta e tamanho da mesa')
sns.scatterplot(x="total_bill", y="tip", size = 'size', sizes=(15, 200), data=tips)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Gorjetas')
sns.scatterplot(x="total_bill", y="tip", size = 'size', hue = 'sex', style = 'smoker', sizes=(15, 200), data=tips)
plt.show()
```
